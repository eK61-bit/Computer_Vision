```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 3.17 - Morfolojik Operatörler -2

tabloResim = cv2.imread('DATA/16_table.png',0)
tabloResim = cv2.bitwise_not(tabloResim)

yatayKernel = np.ones(shape=(1,30))
yatayErosion = cv2.erode(tabloResim,yatayKernel,iterations=1)

dikeyKernel = np.ones(shape=(30,1))
dikeyErosion = cv2.erode(tabloResim,dikeyKernel,iterations=1)

sonuc = tabloResim - yatayErosion - dikeyErosion
sonuc = cv2.bitwise_not(sonuc)
plt.imshow(sonuc,cmap='gray')
```

```python
f, eksen = plt.subplots(1,1,figsize=(15,10))
eksen.imshow(sonuc, cmap="gray")
```

```python
f, eksen = plt.subplots(1,2,figsize=(15,10))
eksen[0].imshow(yatayErosion, cmap="gray")
eksen[1].imshow(dikeyErosion, cmap="gray")
```
