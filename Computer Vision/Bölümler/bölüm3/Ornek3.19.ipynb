```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 3.19 - Histogram Eğrisi
img1 = cv2.imread('DATA/18_jungleOtel.jpg')
img1_B_hist = cv2.calcHist([img1],channels=[0],mask=None,histSize=[256],ranges=[0,256])
img1_G_hist = cv2.calcHist([img1],channels=[1],mask=None,histSize=[256],ranges=[0,256])
img1_R_hist = cv2.calcHist([img1],channels=[2],mask=None,histSize=[256],ranges=[0,256])

imgHSV = cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
imgHSV[:,:,2] = cv2.equalizeHist(imgHSV[:,:,2])
img2 = cv2.cvtColor(imgHSV,cv2.COLOR_HSV2RGB)
img2_R_hist = cv2.calcHist([img2],channels=[0],mask=None,histSize=[256],ranges=[0,256])
img2_G_hist = cv2.calcHist([img2],channels=[1],mask=None,histSize=[256],ranges=[0,256])
img2_B_hist = cv2.calcHist([img2],channels=[2],mask=None,histSize=[256],ranges=[0,256])

f, eksen = plt.subplots(2,3,figsize=(19,7))
eksen[0,0].plot(img1_R_hist)
eksen[0,0].title.set_text("Eşitlemeden Önce R_hist")
eksen[0,1].plot(img1_G_hist)
eksen[0,1].title.set_text("Eşitlemeden Önce G_hist")
eksen[0,2].plot(img1_B_hist)
eksen[0,2].title.set_text("Eşitlemeden Önce B_hist")
eksen[1,0].plot(img2_R_hist)
eksen[1,0].title.set_text("Eşitlemeden Sonra R_hist")
eksen[1,1].plot(img2_G_hist)
eksen[1,1].title.set_text("Eşitlemeden Sonra G_hist")
eksen[1,2].plot(img2_B_hist)
eksen[1,2].title.set_text("Eşitlemeden Sonra B_hist")

img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
f2, eksen2 = plt.subplots(1,2,figsize=(19,7))
eksen2[0].imshow(img1)
eksen2[0].title.set_text("Eşitlemeden Önce")
eksen2[1].imshow(img2)
eksen2[1].title.set_text("Eşitlemeden Sonra")
```
