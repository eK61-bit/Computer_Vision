```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Ã–rnek 3.9 - blurring

resim1 = cv2.imread('DATA/6_noised_pepber.jpg',0)

sonuc1 = cv2.blur(resim1,ksize=(3,3))
sonuc2 = cv2.medianBlur(resim1,ksize=3)
sonuc3 = cv2.GaussianBlur(resim1,ksize=(3,3),sigmaX=50)
sonuc4 = cv2.bilateralFilter(resim1,d=3,sigmaColor=20,sigmaSpace=20)
sonuc5 = cv2.boxFilter(resim1,ddepth=-1,ksize=(3,3))
```

```python
pencere = plt.figure(figsize=(15,10))
pencere.add_subplot(2,3,1)
plt.imshow(resim1,cmap='gray')
pencere.add_subplot(2,3,2)
plt.imshow(sonuc1,cmap='gray')
pencere.add_subplot(2,3,3)
plt.imshow(sonuc2,cmap='gray')
pencere.add_subplot(2,3,4)
plt.imshow(sonuc3,cmap='gray')
pencere.add_subplot(2,3,5)
plt.imshow(sonuc4,cmap='gray')
pencere.add_subplot(2,3,6)
plt.imshow(sonuc5,cmap='gray')
```
