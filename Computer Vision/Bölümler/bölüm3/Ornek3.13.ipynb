```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 3.14 - resim ekleme/birleştirme/çıkarma

x = cv2.imread('DATA/3_opencv-logo.png')
y = cv2.imread('DATA/4_python_logo.png')
x = cv2.cvtColor(x,cv2.COLOR_BGR2RGB)
y = cv2.cvtColor(y,cv2.COLOR_BGR2RGB)
y = cv2.resize(y,(x.shape[1],x.shape[0]))

sonuc1 = x + y
sonuc2 = cv2.add(x,y)
sonuc3 = cv2.addWeighted(x,0.2,y,0.8,gamma=10)
sonuc4 = cv2.subtract(x,y)

f, eksen = plt.subplots(1,4)
eksen[0].imshow(sonuc1, cmap="gray")
eksen[1].imshow(sonuc2, cmap="gray")
eksen[2].imshow(sonuc3, cmap="gray")
eksen[3].imshow(sonuc4, cmap="gray")
```
