```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 3.7 - adaptiveThreshold thresholding

orijinalResim = cv2.imread('DATA/5_gazete.png',0)
sonuc1 = cv2.adaptiveThreshold(orijinalResim,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,55,1)
sonuc2 = cv2.adaptiveThreshold(orijinalResim,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,55,1)
sonuc3 = cv2.adaptiveThreshold(orijinalResim,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,55,1)

f, eksen = plt.subplots(2,2)
eksen[0,0].imshow(orijinalResim, cmap="gray")
eksen[0,1].imshow(sonuc1, cmap="gray")
eksen[1,0].imshow(sonuc2, cmap="gray")
eksen[1,1].imshow(sonuc3, cmap="gray")
```
