```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 3.11 - kernel tabanlı filtreleme

resim = cv2.imread('DATA/8_pepber.jpg',0)
sonuc1 = resim.copy()
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

for k in range(1,resim.shape[1]-2):
    for m in range(1,resim.shape[0]-2):
        sonuc1[k,m] = (resim[k - 1][m - 1] * kernel[0,0] +
                resim[k - 1][m] * kernel[0,1] +
                resim[k - 1][m + 1] * kernel[0,2] +
                resim[k][m - 1] * kernel[1,0] +
                resim[k][m] * kernel[1,1] +
                resim[k][m + 1] * kernel[1,2] +
                resim[k + 1][m - 1] * kernel[2,0] +
                resim[k + 1][m] * kernel[2,1] +
                resim[k + 1][m + 1] * kernel[2,2])
        if sonuc1[k,m] < 0:
            sonuc1[k,m] = 0

sonuc2 = cv2.filter2D(resim,-1,kernel)

f, eksen = plt.subplots(1,3)
eksen[0].imshow(resim, cmap="gray")
eksen[1].imshow(sonuc1, cmap="gray")
eksen[2].imshow(sonuc2, cmap="gray")
```
