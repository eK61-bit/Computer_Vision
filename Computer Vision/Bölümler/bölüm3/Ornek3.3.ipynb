```python
#Örnek 3.3 – Çizim Fonksiyonları

import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

bos_resim = np.zeros(shape=(512,512,3))

cv2.rectangle(bos_resim,pt1=(400,200),pt2=(200,0),color=(0,0,255), thickness=10)

cv2.circle(img=bos_resim,center=(100,300),radius=50,color=(0,255,0), thickness=-1)

cv2.line(bos_resim,pt1=(0,42),pt2=(470,512),color=(255,255,255),thickness=10)

cv2.putText(bos_resim, text='Selam', org=(10,500), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,
            fontScale=5, color=(250,150,35), thickness=3)

plt.imshow(bos_resim)
```
