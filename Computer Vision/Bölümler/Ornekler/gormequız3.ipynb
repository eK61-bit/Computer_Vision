```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
bosresim = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(bosresim)
```

```python
cv2.rectangle(bosresim, pt1=(250,250), pt2=(200,200), color = (0,0,255), thickness=10)                   #mavi dikdörtgen
cv2.rectangle(bosresim, pt1=(500,181), pt2=(332,332), color = (0,255,0), thickness=10)                   #yesil dikdörtgen
cv2.circle(bosresim, center=(300,200), radius=50, color=(255,0,0), thickness=10)                         #kırmızı daire
cv2.circle(bosresim, center=(400,100), radius=50, color=(255,255,0), thickness=-1)                       #sarı daire
cv2.line(bosresim, pt1=(0,0), pt2=(512,512), color=(255,0,255), thickness=5)                             #mor cizgi
cv2.putText(bosresim, "Merhaba",(128,500),cv2.FONT_HERSHEY_COMPLEX,2,color=(255,255,255),thickness=2)    #beyaz yazı
plt.imshow(bosresim)
```
