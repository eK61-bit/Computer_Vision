```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
resim = cv2.imread('DATA/00-cat.jpg')
```

```python
plt.imshow(resim)
```

```python
duzresim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)
griresim = cv2.cvtColor(resim, cv2.IMREAD_GRAYSCALE)
yeniresim = cv2.resize(duzresim,(1000,200))
tersresim = cv2.flip(duzresim, -1)


f,eksen = plt.subplots(2,2)
eksen[0][0].imshow(duzresim)
eksen[0][1].imshow(griresim)
eksen[1][0].imshow(yeniresim)
eksen[1][1].imshow(tersresim)
```
