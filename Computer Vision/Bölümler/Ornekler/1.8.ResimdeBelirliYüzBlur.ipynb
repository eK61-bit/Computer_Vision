```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

buyuk = cv2.imread('../DATA/but2Buyuk.jpg')
buyuk = cv2.cvtColor(buyuk, cv2.COLOR_BGR2RGB)

mert = cv2.imread('../DATA/but2Kucuk.jpg')
mert = cv2.cvtColor(mert, cv2.COLOR_BGR2RGB)

h, w, c = mert.shape

sonuc = cv2.matchTemplate(buyuk,mert,cv2.TM_CCOEFF)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(sonuc)


startLoc = maxLoc
stopLoc = (startLoc[0]+w,startLoc[1]+h)
cv2.rectangle(buyuk,startLoc,stopLoc,(255,255,0),5)

top_left = maxLoc
bottom_right = (top_left[0] + w, top_left[1] + h)
buyuk[top_left[1] : bottom_right[1], top_left[0] : bottom_right[0]] = cv2.medianBlur(buyuk[top_left[1] : bottom_right[1], top_left[0] : bottom_right[0]], 45)


plt.figure(figsize=(16,8))
plt.imshow(buyuk)
```
