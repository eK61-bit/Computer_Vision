```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
orjinalResim = cv2.imread('DATA/12_normal_gray.jpg')
noised = cv2.imread('DATA/13_noised.jpg')

backGroundNoise = cv2.imread('DATA/14_backGround_noise.jpg')
foreGroundNoise = cv2.imread('DATA/15_foreGround_noise.jpg')

kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(7,7))

#Sütun 1
erosionSonuc1 = cv2.erode(orjinalResim, kernel, 3)
dilationSonuc1 = cv2.dilate(orjinalResim, kernel, 3)
#Sütun 2
erosionSonuc2 = cv2.erode(orjinalResim, kernel, 2)
dilationSonuc2 = cv2.dilate(orjinalResim, kernel, 2)
#Sütun 3
erosionSonuc3 = cv2.erode(backGroundNoise, kernel, 1)
openingSonuc = cv2.morphologyEx(backGroundNoise, cv2.MORPH_OPEN, kernel)
#Süton 4
dilationSonuc3 = cv2.dilate(foreGroundNoise, kernel, 1)
closingSonuc = cv2.morphologyEx(foreGroundNoise, cv2.MORPH_CLOSE, kernel)
#Sütun 5
gradientSonuc = cv2.morphologyEx(orjinalResim, cv2.MORPH_GRADIENT, kernel)
topHatSonuc = cv2.morphologyEx(backGroundNoise, cv2.MORPH_TOPHAT, kernel)
blackHatSonuc = cv2.morphologyEx(foreGroundNoise, cv2.MORPH_BLACKHAT, kernel)
```

```python
f, eksen = plt.subplots(3,4)
eksen[0][0].imshow(erosionSonuc1)
eksen[0][1].imshow(dilationSonuc1)
eksen[0][2].imshow(erosionSonuc2)
eksen[0][3].imshow(dilationSonuc2)
eksen[1][0].imshow(erosionSonuc3)
eksen[1][1].imshow(openingSonuc)
eksen[1][2].imshow(dilationSonuc3)
eksen[1][3].imshow(closingSonuc)
eksen[2][0].imshow(gradientSonuc)
eksen[2][1].imshow(topHatSonuc)
eksen[2][2].imshow(blackHatSonuc)
```
