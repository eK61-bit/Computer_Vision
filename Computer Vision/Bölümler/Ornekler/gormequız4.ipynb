```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
x = cv2.imread('DATA/00-cat.jpg')
y = cv2.imread('DATA/01-parrot.jpg')
x = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
y = cv2.cvtColor(y, cv2.COLOR_BGR2RGB)
x = cv2.resize(x,(1000,1000))
y = cv2.resize(y,(1000,1000))
```

```python
sonuc1 = x + y
sonuc2 = cv2.add(x,y)
sonuc3 = cv2.addWeighted(x, 0.9, y, 0.5, gamma=10)
sonuc4 = cv2.subtract(x,y)
```

```python
f,eksen = plt.subplots(2,2)
eksen[0][0].imshow(sonuc1)
eksen[0][1].imshow(sonuc2)
eksen[1][0].imshow(sonuc3)
eksen[1][1].imshow(sonuc4)
```

```python
resim = cv2.imread('DATA/02_animal_in_backpack.jpg')
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

yeniresim = resim.copy()
for yukseklik in range(resim.shape[0]):
    for genislik in range(resim.shape[1]):
        for kanal in range(resim.shape[2]):
            yeniresim[yukseklik, genislik, kanal] = np.clip(1.3 * resim[yukseklik, genislik, kanal] + 15, 0, 255)

f,eksen = plt.subplots(1,2)
eksen[0].imshow(resim)
eksen[1].imshow(yeniresim)
```
