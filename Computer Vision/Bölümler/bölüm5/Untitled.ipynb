```python
import cv2
import time

# Örnek 5.2: OpenCV ile Video Kaydetme

video = cv2.VideoCapture("videoplayback.mp4")

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = video.get(cv2.CAP_PROP_FPS)

if video.isOpened() == False:
    print("Doğru gitmeyen bir şeyler var.")
    
def damga_bas(damga_bas,x,y):
    
    global center,clicked
    
    if damga_bas == cv2.EVENT_LBUTTONDOWN:
        
        center = (x,y)
        clicked = False
        
center = (0,0)
clicked = False

cv2.setMouseCallback('Test', damga_bas) 

while video.isOpened:
    ret, frame = video.read()

    if ret == True:
        time.sleep(1 / fps)
        
        if clicked == True:
            
            kucukResim = cv2.imread('DATA/11_opencv-logo-2.png')
            buyukResim = frame
            kucukResim = cv2.cvtColor(kucukResim,cv2.COLOR_BGR2RGB)#bunları silersek video normal renginde olur
            buyukResim = cv2.cvtColor(buyukResim,cv2.COLOR_BGR2RGB)#bunları silersek video normal renginde olur

            kYukseklik,kGenislik,kKanal = kucukResim.shape
            bYukseklik,bGenislik,bKanal = buyukResim.shape

            roi_alan = buyukResim[bYukseklik-kYukseklik:, bGenislik-kGenislik:].copy()
            kucukResimGray = cv2.cvtColor(kucukResim,cv2.COLOR_RGB2GRAY)

            ret, maske = cv2.threshold(kucukResimGray, 10, 255, cv2.THRESH_BINARY)
            maske_tersi = cv2.bitwise_not(maske)

            roi_alan_bg = cv2.bitwise_and(roi_alan, roi_alan, mask = maske_tersi)
            roi_alan_fg = cv2.bitwise_and(kucukResim, kucukResim, mask = maske)

            birlesmisParca = cv2.add(roi_alan_bg, roi_alan_fg)

            buyukResim[bYukseklik-kYukseklik:, bGenislik-kGenislik:] = birlesmisParca
            
            
        

        cv2.imshow('Video Dosya Okuma', frame)
        

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    else:
        break


video.release()
cv2.destroyAllWindows()
```

```python
import cv2
import time

# mouse callback function
def draw_stamp(event,x,y,flags,param):
    global center,clicked
    
    # get mouse click on down and track center
    if event == cv2.EVENT_LBUTTONDOWN:
        center = (x, y)
        clicked = False
    
    # Use boolean variable to track if the mouse has been released
    if event == cv2.EVENT_LBUTTONUP:
        clicked = True

# Load stamp image
stamp = cv2.imread("DATA/")

# Get stamp size
stamp_height, stamp_width = stamp.shape[:2]

# Haven't drawn anything yet!
center = (0,0)
clicked = False

# Capture Video
cap = cv2.VideoCapture("DATA/videoplayback_4.mp4") 

# Create a named window for connections
cv2.namedWindow('Test')

# Bind draw_stamp function to mouse clicks
cv2.setMouseCallback('Test', draw_stamp) 

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Use if statement to see if clicked is true
    if clicked==True:
        # Get top left corner coordinates of stamp
        top_left_x = center[0] - stamp_width // 2
        top_left_y = center[1] - stamp_height // 2
        
        # Get bottom right corner coordinates of stamp
        bottom_right_x = center[0] + stamp_width // 2
        bottom_right_y = center[1] + stamp_height // 2
        
        # Overlay stamp image on frame
        frame[top_left_y:top_left_y+stamp_height, top_left_x:top_left_x+stamp_width] = stamp
        
    # Display the resulting frame
    cv2.imshow('Test', frame)

    # This command let's us quit with the "q" button on a keyboard.
    # Simply pressing X on the window won't work!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```
