```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Örnek 4.13: Template Matching (Şablon Eşleştirme)

searchIMG = cv2.imread('DATA/28_animalsPiece.jpg')
targetIMG = cv2.imread('DATA/27_animals.jpg')

resultMap = cv2.matchTemplate(targetIMG,searchIMG,cv2.TM_CCOEFF)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(resultMap)

w,h,c = searchIMG.shape

startLoc = maxLoc
stopLoc = (startLoc[0]+w,startLoc[1]+h)

cv2.rectangle(targetIMG,startLoc,stopLoc,(255,255,0),5)

f, eksen = plt.subplots(1,3,figsize=(20,5))
eksen[0].imshow(resultMap)
eksen[1].imshow(resultMap,cmap='gray')
eksen[2].imshow(targetIMG)
```
