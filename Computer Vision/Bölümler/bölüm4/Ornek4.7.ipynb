```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# rnek 4.7: Erosion ile Kenar Tespiti

img = cv2.imread('DATA/16_table.png',0)
img = cv2.bitwise_not(img)

horizontalKernel = np.ones(shape=(1,20))
horizontalErosion = cv2.erode(img,horizontalKernel,iterations=1)

verticalKernel = np.ones(shape=(30,1))
verticalErosion = cv2.erode(img,verticalKernel,iterations=1)

result = img - horizontalErosion - verticalErosion

f, eksen = plt.subplots(2,2,figsize=(17,7))
eksen[0,0].imshow(img,cmap="gray")
eksen[0,1].imshow(horizontalErosion,cmap="gray")
eksen[1,0].imshow(verticalErosion,cmap="gray")
eksen[1,1].imshow(result,cmap="gray")
```
