```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1: Resim verip bu resmin köşelerinden 100 er pixel aralıklarla çerçeve oluşturun ?

# Çözüm 1

resim = cv2.imread("../DATA/00-puppy.jpg")
resim = cv2.cvtColor(resim, cv2.COLOR_BGR2RGB)

def resmi_göster(resim):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(resim)

```

```python
resmi_göster(resim)
print(resim.shape)
```

```python
cerceve_rengi = [255,255,255]
cerceve = cv2.copyMakeBorder(resim,100,100,100,100,cv2.BORDER_CONSTANT, value = cerceve_rengi)
resmi_göster(cerceve)
print(cerceve.shape)
```

```python
# Çözüm 2

resim_copy = resim.copy()
resim_copy = cv2.resize(resim_copy, dsize = (2050,1400))
köseler = np.array([[100,100],[100,1300],[1950,1300],[1950,100]])
pts = köseler.reshape((-1,1,2))
cerceve = cv2.polylines(resim_copy,[pts],isClosed = True, color = (0,0,0), thickness=150)
resmi_göster(cerceve)

```

```python
resim1 = cv2.imread("1_ayasofya.jpg")
resim2 = cv2.imread("11_opencv-logo-2.png")
resim1 = cv2.cvtColor(resim1, cv2.COLOR_BGR2RGB)
resim2 = cv2.cvtColor(resim2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(resim1)
```

```python
print(resim2.shape)
print(resim1.shape)

resim1 = cv2.resize(resim1, dsize = (800,800))
resim2 = cv2.resize(resim2, dsize = (300,300))
print(resim2.shape)
print(resim1.shape)
```

```python
x = 800-300
y = 0
x_son = x + resim2.shape[1]
y_son = y + resim2.shape[0]

sag_üst = resim1[y:y_son, x:x_son]

gri_resim = cv2.cvtColor(resim2, cv2.COLOR_BGR2GRAY)
ret, maske = cv2.threshold(gri_resim,0,255,cv2.THRESH_BINARY)
ters_maske = cv2.bitwise_not(maske)

resim = cv2.bitwise_or(sag_üst,sag_üst, mask = ters_maske)
alan = cv2.bitwise_or(resim2, resim2, mask = maske)

birlestir = cv2.add(resim, alan)

resim1[y:y_son, x:x_son] = birlestir
plt.imshow(resim1)
```
