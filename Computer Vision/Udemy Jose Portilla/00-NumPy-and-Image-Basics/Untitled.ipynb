```python
bos_resim = np.zeros(shape=(512,512,3),dtype = int)
bos_resim.shape
```

```python
plt.imshow(bos_resim)
```

```python
# Dikdörtgen
bos_resim_copy = bos_resim.copy()
cv2.rectangle(bos_resim_copy, pt1 = (100,100),pt2=(300,500),color = (255,0,0),thickness=4)
plt.imshow(bos_resim_copy)
```

```python
cv2.putText(bos_resim_copy, text = "Hello", org = (100,400), fontFace = cv2.FONT_HERSHEY_COMPLEX, fontScale = 4, thickness =5, color = (100,100,100) )
plt.imshow(bos_resim_copy)
```
