```python
# Canny Algoritması Tüm Videoda

import cv2
import time
import numpy as np
cam = cv2.VideoCapture('videoplayback.mp4')

if cam.isOpened() == False:
    print("Doğru gitmeyen bir şeyler var.")
fSayisi=0
fps=cam.get(cv2.CAP_PROP_FPS)
saniye=0
while cam.isOpened:
    ret, frame = cam.read()
    fSayisi+=1
    frame = cv2.Canny(frame,40,40)
    medianValue = np.median(frame)
    if ret==True:
        cv2.imshow('Video Dosya Okuma',frame)
                
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    else:
        break

cam.release()
cv2.destroyAllWindows()
```

```python
# Herhangi bir saniyede transparan çıkan logo


import cv2
import time
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
video=cv2.VideoCapture('videoplayback.mp4')
foto=cv2.imread('11_opencv-logo-2.png')

yuk , gen , kan = foto.shape
fps=video.get(cv2.CAP_PROP_FPS)
#width=int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
#height=int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
#vw=cv2.VideoWriter('aaa25incikare.mp4',cv2.VideoWriter_fourcc(*'XVID'),fps,(width,height),isColor=True)

fSayisi=0


kucukResimGray = cv2.cvtColor(foto,cv2.COLOR_RGB2GRAY)
ret, maske = cv2.threshold(kucukResimGray, 10, 255, cv2.THRESH_BINARY)
maske_tersi = cv2.bitwise_not(maske)



if video.isOpened()==False:
    print("noo")
    
while video.isOpened():

    
    ret,frame=video.read()
    fyuk,fgen,fkanal=frame.shape
    fSayisi=fSayisi+1
    bbaspix=int((fyuk-yuk)/2)     
    bbitpix=bbaspix+yuk     
    ebaspix=int((fgen-gen)/2)     
    ebitpix=ebaspix+gen
    
    roi_alan = frame[bbaspix:bbitpix,ebaspix:ebitpix].copy()
        
    time.sleep(1/fps)

    roi_alan_bg = cv2.bitwise_and(roi_alan, roi_alan, mask = maske_tersi)
    roi_alan_fg = cv2.bitwise_and(foto, foto, mask = maske)
    
        
    birlesmisParca = cv2.add(roi_alan_bg, roi_alan_fg)
    
    if ret==True:
        if fSayisi/fps>2 and fSayisi/fps<5:
            frame[bbaspix:bbitpix,ebaspix:ebitpix] = birlesmisParca
        
        #cv2.putText(frame,str(height),(500,300),cv2.FONT_HERSHEY_COMPLEX,3,(255,0,0))
        
        cv2.imshow('a',frame)
        #vw.write(frame)
        
        if cv2.waitKey(10)&0XFF==27:
            break
            
    
video.release()
#vw.release()
cv2.destroyAllWindows()
```
